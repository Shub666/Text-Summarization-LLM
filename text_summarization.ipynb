### Prompt template text summarization

```python
from langchain_classic.chains import LLMChain
from langchain_classic import PromptTemplate

genericprompt="""
write a summery of the following speech:
Speech:{speech}
Translate the precise summery to {language}
"""

prompt=PromptTemplate(
    input_variables=['speech','language'],
    template=genericprompt
)
prompt
```

```python
complete_prompt=prompt.format(speech=speech,language="French")
```

```python
llm.get_num_tokens(complete_prompt)
```

```python
llm_chain = LLMChain(llm=llm,prompt=prompt)
summery=llm_chain.run({'speech':speech,'language':'hindi'})
summery 
```

### Stuff Document Chain Text summarization


```python
## Stuff Document Chain Text summarization

from langchain_community.document_loaders import PyPDFLoader
loader= PyPDFLoader("apjspeech.pdf")
docs= loader.load_and_split()
docs
```

```python
template="""
Write a concise and short summery of the fallowing speech
Speech:{text}

"""

prompt=PromptTemplate(input_variable=['text'],template=template)
prompt
```

```python
from langchain_classic.chains.summarize import load_summarize_chain

chain=load_summarize_chain(llm=llm,chain_type='stuff',prompt=prompt,verbose=True)
output_summery=chain.run(docs)
output_summery
```

### MAP Reduce Text summarization

```python
from langchain_classic.text_splitter import RecursiveCharacterTextSplitter

docs
```

```python
final_doc= RecursiveCharacterTextSplitter(chunk_size=2000,chunk_overlap=100).split_documents(docs)
final_doc
```

```python
len(final_doc)
```

```python
prompt
```

```python
chunk_prompt="""
write a summery of the below speech:
Speech:{text}
Summery:
"""

map_prompt_template=PromptTemplate(input_variable=['text'],template=chunk_prompt)
```

```python
final_prompt="""
Provide the final summery of the entire speech with below points.
Add a Motivation tile,Start the precise summery with an intro and provide the summery in numbers
points for the speech
Speech:{text}
"""
final_prompt_template=PromptTemplate(
    llm=llm,
    template=final_prompt
)
```

```python
summery_chain = load_summarize_chain(
    llm=llm,
    chain_type="map_reduce",
    map_prompt=map_prompt_template,
    combine_prompt=final_prompt_template,
    verbose=True
)

outout= summery_chain.run(final_doc)
outout
```

### Refine Chain summarization

```python
chain = load_summarize_chain(
    llm=llm,
    chain_type="refine",
    verbose=True
)

output_summery = chain.run(final_doc)
output_summery
```
